```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
output = classifier('It is okay we are just friend')
```

```python
classifier('It is okay we are just friend')
```

```python
print(f"The sentiment is : {output[0]['label']}")
print(f"The sentiment is : {output[0]['score']}")
```

```python
multiple_lines = classifier([
    'Hi we are so happy to see you here!',
    "We hope you dont hate it"
])
```

```python
for i, sentiment in enumerate(multiple_lines):
  print(f"For line number : {i+1}")
  print(f"The sentiment is : {sentiment['label']}")
  print(f"The sentiment is : {sentiment['score']}")
  print()
```
